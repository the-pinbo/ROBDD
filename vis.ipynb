```python
import pydot
import os
```

```python
rootDir = r"C:\Users\inba2\Documents\IEEEMedia"
G = pydot.Dot(graph_type="digraph")
node = pydot.Node(rootDir.split("\\")[-1],style="filled",fillcolor="green")
G.add_node(node)
```

```python
for root, dirs,files in os.walk(rootDir):
    # print(root)
    # print(root.split("\\")[-1])
    for subdir in dirs:
        node = pydot.Node(subdir,style="filled",fillcolor="yellow")
        G.add_node(node)
        edge = pydot.Edge(root.split("\\")[-1],subdir)
        G.add_edge(edge)
    for f in files:
        node = pydot.Node(f,style="filled",fillcolor="orange")
        G.add_node(node)
        edge = pydot.Edge(root.split("\\")[-1],f)
        G.add_edge(edge)
```

```python
from IPython.display import Image, display
img = Image(G.create_png())
display(img)
```
